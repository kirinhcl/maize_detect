```python

import glob
import os
from PIL import Image

output_path = "/Users/luchenghao/Downloads/maize/rotate"

img_list = []
img_list.extend(glob.glob('*.jpg'))
img_list.extend(glob.glob('*.JPG'))

print(img_list)

for img_path in img_list:
    img_name = os.path.splitext(img_path)[0]
    print(img_name)
    im=Image.open(img_path)
    im=im.convert("RGB")
    im_rotate=im.rotate(90,expand=1)
    folder=os.path.exists(output_path)
    if not folder:
        os.makedirs(output_path)
        
    im_rotate.save(output_path+'/'+img_name+'_rotate'+'.jpg')
    
print('finished')
    
```
